```python
# Account 클래스 생성 후 생성자를 구현, 생성자에서는 예금주와 초기 잔액만 입력 받는다.
# 은행이름은 SC은행, 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성 ex. 111-11-111111
import random




class Account:


  product_account_num = 0
  #클래스 변수로 생성된 계좌 객체 개수 추적

  #생성자 생성, 이름과 초기 잔액을 입력받기
  #은행이름과 계좌번호 생성
  def __init__(self, name, deposit) :
    self.name = name
    self.deposit = deposit
    Account.product_account_num += 1
    #생성된 계좌 객체 개수 추가

    self.bank = "SC은행"      #은행명
    account_num1 = random.randint(0,999)   #계좌번호 랜덤 생성
    account_num2 = random.randint(0,99)    #계좌번호 랜덤 생성
    account_num3 = random.randint(0,999999)#계좌번호 랜덤 생성

    account_num1 = str(account_num1).zfill(3)        #계좌번호 랜덤 생성 후 빈곳 0으로 채우기, 문자열 0으로 채우기 검색해서 참조
    account_num2 = str(account_num2).zfill(2)        #계좌번호 랜덤 생성 후 빈곳 0으로 채우기
    account_num3 = str(account_num3).zfill(6)        #계좌번호 랜덤 생성 후 빈곳 0으로 채우기
    self.account_num = account_num1+"-"+account_num2+"-"+account_num3   #계좌번호 생성 완료

    self.deposit_count = 0  #입금 횟수 확인

    self.deposit_his = []            #입금 거래 내역 리스트용
    self.withdraw_his = []           #출금 거래 내역 리스트용

  def get_account_num(self) :
    print(f"지금까지 생성된 계좌의 개수는 : {Account.product_account_num}개 입니다.")
  #생성된 계좌 객체 개수 가져와서 출력

  #입금을 위한 deposit 메서드, 입금은 1원 이상,
  def money_deposit(self) :
    self.m_plus = 0
    self.m_plus = int(input("입금하실 금액을 입력해주세요 : "))

    if self.m_plus >= 1 :  #1원 이상 입금받기

      self.deposit_his.append(self.m_plus) #입금 받은 금액을 입금 내역에 추가
      self.deposit += self.m_plus   #잔고에 입금 받은 금액 추가
      self.deposit_count += 1     #입금 횟수 추가
      print(f"{self.m_plus}원 입금되었습니다.")
      print(f"현재 잔고는 {self.deposit}원 입니다.")

      if self.deposit_count % 5 == 0 :  #5의 배수 5회마다
        self.deposit = self.deposit * 1.01 #잔고의 1퍼센트 이자 지급


    else :    #입금하는 금액과 잔고 출력
      print("입금 금액은 1원 이상부터 가능합니다.")

  #출금을 위한 withdraw 메서드, 출금은 계좌의 잔고이상은 불가능
  def money_withdraw(self) :
    self.m_minus = 0
    self.m_minus = int(input("출금하실 금액을 입력해주세요 : "))

    if self.deposit < self.m_minus : #잔고보다 출금 금액이 클경우
      print("계좌의 잔고보다 많은 금액으로 출금할 수는 없습니다.")

    else :  #출금하는 금액과 잔고 출력
      self.deposit -= self.m_minus         #잔고에서 출금 금액 빼기
      self.withdraw_his.append(self.m_minus)     #출금 금액을 출금 내역에 추가
      print(f"{self.m_minus}원 출금되었습니다.")
      print(f"현재 잔고는 {self.deposit}원 입니다.")

#계좌 정보 가져오기
  def display_info(self) :
    print("은행이름 : ",self.bank)
    print("예금주 : ",self.name)
    print("계좌번호 : ",self.account_num)
    print("잔고 : " f"{self.deposit:,}")   #3자리마다 ,표시가능


  def deposit_history(self) :
    for x in self.deposit_his:
      print(x)
    #입금 거래 내역 앞에서부터 쫘아아아악

  def withdraw_history(self) :
    for y in self.withdraw_his:
      print(y)
    #출금 거래 내역 앞에서부터 쭈와아아아아아악

data = []

kang = Account("강임구",2200000) #개체1
park = Account("박민천",1450000) #개체2
lee = Account("이남석", 4150000) #개체3

data.append(kang)
data.append(park)
data.append(lee)

#추가한 객체를 data 리스트에 저장

#data 리스트안 객체중에 잔고가 100만원 이상인 계좌정보 출력
for i in data:
  if i.deposit >= 1000000:
    i.display_info()



# kang.get_account_num()
# park.get_account_num()
# kang.money_deposit()
# kang.money_withdraw()
# kang.display_info()
# kang.deposit_history()
# kang.withdraw_history()
```
